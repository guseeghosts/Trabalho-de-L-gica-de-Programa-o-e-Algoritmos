```python
#Mensagem de boas-vindas
print('Bem vindo a loja de Gustavo Queiroz!')

#Entrada do valor unitário e quantidade do produto
valor = float(input('Digite o valor do produto: '))
qtd = int(input('Digite a quantidade do produto: '))

#Cálculo do valor total da compra
total = valor * qtd

#Aplicação do desconto de acordo com as condições
if total >= 2500 and total < 6000:
  desconto = 4
elif total >= 6000 and total < 10000:
  desconto = 7
elif total >= 10000:
  desconto = 11
else:
  desconto = 0

#Cálculo do valor do desconto e o valor final, com desconto
valor_desconto = total * (desconto / 100)
valor_final = total - valor_desconto

#Print dos valores da compra, desconto e valor total com desconto, respectivamente
print(f'Valor total SEM desconto: R$ {total:.2f}')
print(f'Valor do desconto: R$ {valor_desconto:.2f}')
print(f'Valor total COM desconto: R$ {valor_final:.2f}')
```

```python
#Mensagem de boas-vindas com cardápio
print('       Bem vindo a Gelateria de Gustavo Queiroz!       ')
print('-' * 23,'Cardápio','-' * 23)
print('-' * 56)
print('----|  Tamanho  |   Cupuaçu (CP)   |   Açaí (AC)   |----')
print('----|     P     |      R$9.00      |    R$11.00    |----')
print('----|     M     |      R$14.00     |    R$16.00    |----')
print('----|     G     |      R$18.00     |    R$20.00    |----')
print('-' * 56)

soma = 0 #Acumulador da soma dos valores dos pedidos

#Laço de checagem de validez do sabor
while True:
  sabor = input('Digite o sabor desejado: (Cupuaçu (CP) ou (Açaí (AC)): ').upper()
  if sabor not in ['CP', 'AC']:
    print('Sabor inválido!\n')
    continue

#Laço de checagem de validez do tamanho
  while True:
    tamanho = input('Digite o tamanho desejado: (P, M ou G): ').upper()
    if tamanho not in ['P', 'M', 'G']:
      print('Tamanho inválido!\n')
      continue
    break

#Ligação de sabor e tamanho com modelo aninhado de if/elif/else
  if sabor == 'CP':
    if tamanho == 'P':
      soma += 9
      print('Tamanho: P de Cupuaçu - R$9.00\n')
    elif tamanho == 'M':
      soma += 14
      print('Tamanho: M de Cupuaçu - R$14.00\n')
    else:
      soma += 18
      print('Tamanho: G de Cupuaçu - R$18.00\n')
  elif sabor == 'AC':
    if tamanho == 'P':
      soma += 11
      print('Tamanho: P de Açaí - R$11.00\n')
    elif tamanho == 'M':
      soma += 16
      print('Tamanho: M de Açaí - R$16.00\n')
    else:
      soma += 20
      print('Tamanho: G de Açaí - R$20.00\n')

#Pergunta de checagem de continuação ou conclusão da compra
  pergunta = input('Deseja pedir mais alguma coisa? (S/N): ')
  if pergunta == 'S' or  pergunta =='s':
    print(' ')
    continue
  elif pergunta == 'N' or pergunta == 'n':
      print(f'\nValor total é de R$ {soma}') #Print finalizando a compra com o valor total
      break #Quebra do laço
  else:
    print('Opção inválida!') #Verificação de opção inválida para 'S/N'
    continue
```

```python
#Mensagem de boas-vindas com opções de serviços oferecidos, quebra de linha (\n) aplicada
print('Bem vindo a Copiadora de Gustavo Queiroz\n\nServiços oferecidos:\nDIG - Digitalização\nICO - Impressão Colorida\nIPB - Impressão Preto e Branco\nFOT - Fotocópia\n')

#Função para chamar o serviço a escolher
def escolha_servico():
  servico = input('Digite o serviço desejado: ').upper()
  if servico in ['DIG', 'ICO', 'IPB', 'FOT']:
    return servico
  else:
    print('Serviço inválido!\n')
    return escolha_servico()

#Função para chamar as páginas e aplicar os descontos dos intervalos especificados
def num_pagina():
  while True:
    try:
      paginas = int(input('Digite o número de páginas desejado: '))
      if paginas < 20:
        return paginas #0% desconto
      elif paginas >= 20 and paginas < 200:
        return paginas * 0.85 #15% desconto
      elif paginas >= 200 and paginas < 2000:
        return paginas * 0.80 #20% desconto
      elif paginas >= 2000 and paginas < 20000:
        return paginas * 0.75 #25% desconto
      else:
        print('Número de páginas inválido!\n')
    except ValueError:
      print('Número de páginas não aceito!\n')

#Função para chamar a escolha dos serviços opcionais disponíveis
def servico_extra():
  valor_extra = 0
  adicional = input('\nEscolha algum serviço adicional:\n1 - Encadernação Simples\n2 - Encadernação de Capa Dura\n0 - Não desejo adicional\n')
  if adicional == '1':
    valor_extra += 15
  elif adicional == '2':
    valor_extra += 40
  elif adicional == '0':
    return valor_extra
  else:
    print('Opção inválida!')
    return servico_extra()
  return valor_extra

#Coleta os dados da função de escolha de serviço
servico = escolha_servico()

#Atribui os preços a seus respectivos serviços
if servico == 'DIG':
  preco = 1.10
elif servico == 'ICO':
  preco = 1.00
elif servico == 'IPB':
  preco = 0.40
elif servico == 'FOT':
  preco = 0.20

#Coleta os dados da função de escolha de páginas, com seu desconto atribuído
paginas = num_pagina()

#Coleta os dados da função de escolha de possível serviço adicional
adicional = servico_extra()

#Calcula e mostra o valor total a pagar
total = (preco * paginas) + adicional
print(f'O valor total é de R$ {total:.2f}')
```

```python
#Mensagem de boas-vindas da Livraria
print('Bem vindo a Livraria de Gustavo Queiroz')
print('-' * 50)

#Lista dos livros e variável para controle das IDs
lista_livro = []
id_global = 0

#Função para cadastro de livros, cria um dicionário com os dados fornecidos e os adiciona á lista
def cadastrar_livro(id):
  print('-' * 14 + ' MENU CADASTRAR LIVRO ' + '-' * 14 + '\n')
  nome = input('Digite o nome do livro: ')
  autor = input('Digite o autor do livro: ')
  editora = input('Digite a editora do livro: ')
  dicionario = {'id': id, 'nome': nome, 'autor': autor, 'editora': editora}
  lista_livro.append(dicionario)
  print(f'ID do livro: {dicionario["id"]}')

#Função para consultar os livros cadastrados
def consultar_livro():
  while True:
    opcao = int(input('-' * 14 + ' MENU CONSULTAR LIVROS ' + '-' * 14 + '\n' + '1. Consultar Todos\n2. Consultar por Id\n3. Consultar por Autor\n4. Retornar ao menu\n'))
    if opcao == 1:
      if not lista_livro:
        print('Nenhum livro cadastrado!')
        return
      for livro in lista_livro:
        print(f'Id: {livro["id"]}\nNome: {livro["nome"]}\nAutor: {livro["autor"]}\nEditora: {livro["editora"]}\n')
    elif opcao == 2:
      id = int(input('Digite o id do livro: '))
      for livro in lista_livro:
        if livro['id'] == id:
          print(f'Id: {livro["id"]}\nNome: {livro["nome"]}\nAutor: {livro["autor"]}\nEditora: {livro["editora"]}\n')
          break
      else:
        print('Livro não encontrado!')
    elif opcao == 3:
      autor = input('Digite o autor do livro: ')
      encontrado = False
      for livro in lista_livro:
        if livro['autor'] == autor:
          print(f'Id: {livro["id"]}\nNome: {livro["nome"]}\nAutor: {livro["autor"]}\nEditora: {livro["editora"]}\n')
          encontrado = True
      if not encontrado:
        print('Nenhum autor encontrado!')
    elif opcao == 4:
      return
    else:
      print('Opção inválida!')
      continue

#Função para remover livros previamente cadastrados
def remover_livro():
  while True:
    try:
      id = int(input('-' * 15 + ' MENU REMOVER LIVRO ' + '-' * 15 + '\n' + 'Digite o id do livro a ser removido: '))
      for livro in lista_livro:
        if livro['id'] == id:
          lista_livro.remove(livro)
          print('Livro removido com sucesso!')
          return
      print('ID Inválido!')
    except ValueError:
      print('ID Inválido!')

#Código principal, apresentando o menu principal e designando as suas funções
while True:
  try:
    menu_principal = int(input('-' * 17 + ' MENU PRINCIPAL ' + '-' * 17 + '\n' + '1. Cadastrar Livro\n2. Consultar Livro\n3. Remover Livro\n4. Encerrar Programa\n'))
    if menu_principal == 1:
      id_global += 1
      cadastrar_livro(id_global)
    elif menu_principal == 2:
      consultar_livro()
    elif menu_principal == 3:
      remover_livro()
    elif menu_principal == 4:
      print('Programa encerrado!')
      break
  except ValueError:
    print('Opção inválida!')
    continue
```
